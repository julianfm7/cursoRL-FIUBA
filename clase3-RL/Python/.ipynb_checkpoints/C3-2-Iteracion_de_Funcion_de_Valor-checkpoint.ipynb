Vamos a usar la ecuacin de optimalidad para encontrar de manera iterativa la poltica ptima.

```python
# en caso de correrlo en google colab
# de esta manera podremos tener la carpeta lib (donde se encuentra en ambiente Gridworld)

import sys
!git clone https://github.com/julianfm7/cursoRL-FIUBA

# necesario en google colab para que sys.path busque
# y encuentre la carpeta lib donde se encuentra el ambiente Gridworld

!cd cursoRL-FIUBA/

import numpy as np
import pprint

from lib.envs.gridworld import GridworldEnv
```

```python
pp = pprint.PrettyPrinter(indent=2)
env = GridworldEnv()
```

```python
def value_iteration(env, theta=0.0001, discount_factor=1.0):
    """
    Algoritmo de iteracin de la funcin de valor
    
    Argumentos:
        env: ambiente de OpenAI. env.P representa la transicin de probabilidades del ambiente.
            env.P[s][a] es una lista de tuplas de transicin (prob, next_state, reward, done).
            env.nS es el nmero de estados del ambiente.
            env.nA es el nmero de acciones del ambiente.
        theta: Paramos de evaluar una vez que la funcin de valor cambia menos que theta para todos los estados.
        discount_factor: factor de descuento gama.
        
    Retorna:
        Una tupla (policy, V) de poltica ptima y funcin de valor ptima.
    """
    
    # Implementar!

    V = np.zeros(env.nS)
    # primero calcular la funcion valor hasta que el cambio maximo de V sea menor a theta
    while True:
        # por cada estado s
          # calcular el retorno esperado para todas las acciones
          # elegir la accion con mayor retorno
          # mantener una variable (delta) que indica el cambio maximo en V(s) para todos los estados
          # hacer update de V(s) = maximo retorno esperado (mejor accion)
        # si delta < theta: parar
        break
    
    
    policy = np.zeros([env.nS, env.nA])
    # ahora calcular la poltica ptima:
    # por cada estado s
      # calcular el retorno esperado de cada accin a
      # encontrar la accin con mayor retorno esperado a_max
      # la poltica debe realizar esta accin a_max en el estado s con probabilidad 1.0
    
    
    return policy, V
```

```python
policy, v = value_iteration(env)

print("Distribucin de probabilidad de la poltica:")
print(policy)
print("")

print("Poltica mostrada en la grilla: (0=arriba, 1=derecha, 2=abajo, 3=izquierda)")
print(np.reshape(np.argmax(policy, axis=1), env.shape))
print("")

print("Funcin de valor:")
print(v)
print("")

print("Funcin de valor exhibida en la grilla:")
print(v.reshape(env.shape))
print("")
```

```python
# Testear la funcin de valor
expected_v = np.array([ 0, -1, -2, -3, -1, -2, -3, -2, -2, -3, -2, -1, -3, -2, -1,  0])
np.testing.assert_array_almost_equal(v, expected_v, decimal=2)
```
